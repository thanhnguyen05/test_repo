## TASK 2

```python
sns.boxplot(boston_df['MEDV'])
plt.title('Graph Median value of owner-occupied homes')
plt.show()
```

```python
plt.bar(boston_df['CHAS'].unique(), boston_df['CHAS'].value_counts())
plt.xlabel('CHAS')
plt.ylabel('Count')
plt.title('Chas driver')
plt.show()
```

```python
boston_df.loc[boston_df['AGE'] <= 35, 'age_group'] = '35 years and younger age'
boston_df.loc[(boston_df['AGE'] > 35) & (boston_df['AGE'] < 70), 'age_group'] = 'between 35 and 70 years age'
boston_df.loc[boston_df['AGE'] > 70, 'age_group'] = '70 years and older age'
```

```python
ax = sns.boxplot(y='MEDV', data=boston_df, hue='age_group')
```

```python
sns.scatterplot(x='INDUS', y='NOX', data=boston_df)
plt.show()
```

```python
plt.hist(boston_df['PTRATIO'])
plt.xlabel('PTRATIO')
plt.ylabel('COUNT')
plt.show()
```
